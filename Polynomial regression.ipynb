```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Position Salary.csv')
```

```python
df
```

```python
x=df.iloc[:,1:2].values
```

```python
x
```

```python
y=df.iloc[:,2].values
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
sns.lmplot(x='Level',y='Salary',data=df)
```

```python
from sklearn import linear_model
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[6.5]]) # here linear regression is not giving the prediction which we are expecting
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
Poly=PolynomialFeatures(degree=2)
```

```python
x_poly=Poly.fit_transform(x)
```

```python
reg2=linear_model.LinearRegression()
```

```python
reg2.fit(x_poly,y)
```

```python
reg2.predict(Poly.fit_transform([[6.5]]))
```
