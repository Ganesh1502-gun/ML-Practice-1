```python
import pandas as pd
```

```python
df=pd.read_csv('iris_dataset.csv')
```

```python
df
```

```python
df['Species'].unique()
```

```python
df['Species'].replace({'Iris-virginica':'3', 'Iris-setosa':'1', 'Iris-versicolor':'2'},inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']],df['Species'], test_size=0.2) #species are our dependent variable 
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python

lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```

```python
X_test          #{'Iris-virginica':'3', 'Iris-setosa':'1', 'Iris-versicolor':'2'}
```

```python
lr.score(X_test,y_test)
```
