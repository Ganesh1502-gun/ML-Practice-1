```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# basically linear equation wont work as we see too much distance in data point from the line on the graph 
# we need sigmoid fucntion 
# formula for this: y=1/1+e-x (e raise to -x)
# y = dependent v (baught insurance)
# x = independent v (age)
# e = Euler's constant ~ 2.71828
# Sigmoid is used to convert the Independent v into prabality 1 or 0
```

```python
df=pd.read_csv('Terminsurance.csv')
```

```python
df
```

```python
df['Bought Insurance'].replace({'no':'0','yes':'1'},inplace=True)
```

```python
plt.scatter(x='Age',y='Bought Insurance', data=df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['Age']],df['Bought Insurance'],test_size=0.2)
```
